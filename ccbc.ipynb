```python
atk = pkmn['atk']
defesa = pkmn['def']
spatk = pkmn['spatk']
spdef = pkmn['spdef']
spd = pkmn['speed']
off_stats = [atk,defesa,spatk,spdef]
stats = [atk,defesa,spatk,spdef,spd]
```

# Box Plot

```python
fig, ax = plt.subplots()
ax.boxplot(off_stats)
ax.set_title('Boxplot do valor dos Atributos dos Pokémons')
ax.set_xlabel('Tipo do atributo')
ax.set_ylabel('Valor')
xticks = ['Ataque', 'Defesa', 'Ataque Especial', 'Defesa Especial']
ax.set_xticklabels(xticks)
plt.show()
```

# Histograma

```python
fig, ax = plt.subplots()
ax.hist(off_stats)
ax.set_title('Histograma dos Atributos dos Pokémons')
ax.set_xlabel('Valor')
ax.set_ylabel('Quantidade')
plt.show()
```

```python
hei = pkmn['height']
wei = pkmn['weight']
sec_sts = [hei, wei]

fig, ax = plt.subplots()
ax.hist(hei)
ax.set_title('Histograma da Altura dos Pokémons')
ax.set_xlabel('Valor')
ax.set_ylabel('Quantidade')
plt.show()
```
