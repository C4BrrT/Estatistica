```python
atk = pkmn['atk'].value_counts()
defesa = pkmn['def'].value_counts()
spatk = pkmn['spatk'].value_counts()
spdef = pkmn['spdef'].value_counts()
stats = [atk,defesa,spatk,spdef]
```

```python
fig, ax = plt.subplots()
ax.boxplot(stats)
ax.set_title('Boxplot da quantidade dos Atributos dos Pokémons')
ax.set_xlabel('Tipo do atributo')
ax.set_ylabel('Quantidade')
xticks = ['Ataque', 'Defesa', 'Ataque Especial', 'Defesa Especial']
ax.set_xticklabels(xticks)
plt.show()
```
